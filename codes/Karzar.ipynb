```python
import requests  # Import the requests library for making HTTP requests
from bs4 import BeautifulSoup  # Import BeautifulSoup for parsing HTML
from dataclasses import dataclass  # Import dataclass for defining simple data structures
from typing import List, Optional  # Import List and Optional for type hinting
import pandas as pd  # Import pandas for handling data and creating DataFrames

@dataclass
class Campaign:
    """
    A data class to store information about a campaign.
    Each campaign has optional attributes: name, url, signatures, start_date, end_date, and organizer.
    """
    name: Optional[str] = None
    url: Optional[str] = None
    signatures: Optional[str] = None
    start_date: Optional[str] = None
    end_date: Optional[str] = None
    organizer: Optional[str] = None

def fetch_html_content(url: str) -> str:
    """
    Fetch the HTML content of the given URL.
    
    :param url: The URL of the webpage to fetch.
    :return: The HTML content of the page as a string.
    :raises: HTTPError if the HTTP request returned an unsuccessful status code.
    """
    response = requests.get(url)  # Perform a GET request to the given URL
    response.raise_for_status()  # Raise an exception for HTTP errors
    return response.text  # Return the raw HTML content of the page

def safe_extract(element, selector, attribute=None):
    """
    Safely extract text or attribute content from an HTML element.
    
    :param element: The HTML element to search within.
    :param selector: The CSS selector to identify the target element.
    :param attribute: The attribute to extract (if any). If None, extract the text content.
    :return: The extracted content as a string, or None if the element/attribute was not found.
    """
    found = element.select_one(selector)  # Find the first matching element for the given selector
    if found:  # If an element is found
        return found.get(attribute) if attribute else found.get_text(strip=True)  # Return attribute or text
    return None  # Return None if no element is found

def scrape_karzar_campaigns(url: str) -> List[Campaign]:
    """
    Scrape campaign information from the Karzar website.
    
    :param url: The URL of the campaign listing page on Karzar.
    :return: A list of Campaign objects containing the scraped data.
    """
    html_content = fetch_html_content(url)  # Fetch the HTML content of the page
    soup = BeautifulSoup(html_content, 'html.parser')  # Parse the HTML content with BeautifulSoup
    campaign_boxes = soup.select('div.campaign-box-item')  # Select all campaign box items
    
    campaigns = []  # Initialize an empty list to store the campaign data
    for box in campaign_boxes:  # Iterate over each campaign box
        # Extract and store relevant campaign information using the safe_extract function
        campaign = Campaign(
            name=safe_extract(box, 'h3.campaign-box__title'),
            url=safe_extract(box, 'a.campaign-box', 'href'),
            signatures=safe_extract(box, 'span.campaign-box__top-text'),
            start_date=safe_extract(box, 'div.campaign-box__info:contains("شروع کارزار")'),
            end_date=safe_extract(box, 'div.campaign-box__info:contains("پایان کارزار")'),
            organizer=safe_extract(box, 'div.campaign-box__info:contains("از طرف")')
        )
        
        # Data cleaning: Extract and format the relevant parts of the strings
        if campaign.signatures:
            campaign.signatures = campaign.signatures.split()[0]  # Extract the number of signatures
        if campaign.start_date:
            campaign.start_date = campaign.start_date.split(': ')[-1]  # Extract the start date
        if campaign.end_date:
            campaign.end_date = campaign.end_date.split(': ')[-1]  # Extract the end date
        if campaign.organizer:
            campaign.organizer = campaign.organizer.split(': ')[-1]  # Extract the organizer name
        if campaign.url:
            campaign.url = 'https://www.karzar.net' + campaign.url  # Prepend base URL to the relative link
        
        campaigns.append(campaign)  # Add the campaign object to the list
    
    return campaigns  # Return the list of campaigns

def campaigns_to_dataframe(campaigns: List[Campaign]) -> pd.DataFrame:
    """
    Convert a list of Campaign objects into a pandas DataFrame.
    
    :param campaigns: A list of Campaign objects.
    :return: A pandas DataFrame containing the campaign data.
    """
    df = pd.DataFrame([vars(campaign) for campaign in campaigns])  # Convert the list of Campaign objects to a DataFrame
    df.insert(0, 'index', range(1, len(df) + 1))  # Add an index column starting from 1
    return df  # Return the DataFrame

# Example usage
url = "https://www.karzar.net/campaigns?province=2"

try:
    campaigns = scrape_karzar_campaigns(url)  # Scrape the campaigns from the given URL
    df = campaigns_to_dataframe(campaigns)  # Convert the scraped data into a DataFrame

    # Optionally, display the DataFrame or save it to a CSV file
    # print(df)
    # df.to_csv('karzar_campaigns.csv', index=False)

except requests.RequestException as e:  # Handle exceptions related to HTTP requests
    print(f"An error occurred while fetching the URL: {e}")
except Exception as e:  # Handle any other exceptions
    print(f"An error occurred while parsing the content: {e}")

df
```

```python
# Import necessary libraries
import requests  # For making HTTP requests to web pages
from bs4 import BeautifulSoup  # For parsing HTML content
from dataclasses import dataclass  # For creating a data class to store campaign information
from typing import List, Optional, Dict  # For type hinting, making the code more readable and maintainable
import pandas as pd  # For creating and manipulating dataframes
import re  # For using regular expressions (used in creating safe filenames)
import time  # For adding delays between requests

# Define a data class to store information about each campaign
@dataclass
class Campaign:
    # Each field is defined with a type hint and a default value of None
    name: Optional[str] = None  # The name of the campaign
    url: Optional[str] = None  # The URL of the campaign
    signatures: Optional[str] = None  # The number of signatures for the campaign
    start_date: Optional[str] = None  # The start date of the campaign
    end_date: Optional[str] = None  # The end date of the campaign
    organizer: Optional[str] = None  # The organizer of the campaign
    province: Optional[str] = None  # The province where the campaign is active

# Function to fetch HTML content from a given URL
def fetch_html_content(url: str) -> str:
    # Send a GET request to the URL
    response = requests.get(url)
    # Raise an exception for bad status codes (e.g., 404 Not Found)
    response.raise_for_status()
    # Return the HTML content as text
    return response.text

# Function to safely extract data from HTML elements
def safe_extract(element, selector, attribute=None):
    # Find the first element matching the CSS selector
    found = element.select_one(selector)
    if found:
        # If an attribute is specified, return its value
        if attribute:
            return found.get(attribute)
        # Otherwise, return the text content of the element, with whitespace stripped
        else:
            return found.get_text(strip=True)
    # If no element is found, return None
    return None

# Function to scrape campaign data from a single page
def scrape_karzar_campaigns(url: str, province: str) -> List[Campaign]:
    # Fetch the HTML content of the page
    html_content = fetch_html_content(url)
    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')
    # Find all campaign box elements on the page
    campaign_boxes = soup.select('div.campaign-box-item')
    
    # List to store all campaigns found on this page
    campaigns = []
    
    # Iterate through each campaign box
    for box in campaign_boxes:
        # Create a Campaign object for each campaign box
        campaign = Campaign(
            # Extract each piece of information using the safe_extract function
            name=safe_extract(box, 'h3.campaign-box__title'),
            url=safe_extract(box, 'a.campaign-box', 'href'),
            signatures=safe_extract(box, 'span.campaign-box__top-text'),
            start_date=safe_extract(box, 'div.campaign-box__info:contains("شروع کارزار")'),
            end_date=safe_extract(box, 'div.campaign-box__info:contains("پایان کارزار")'),
            organizer=safe_extract(box, 'div.campaign-box__info:contains("از طرف")'),
            province=province
        )
        
        # Clean up the extracted data
        if campaign.signatures:
            # Keep only the number part of the signatures
            campaign.signatures = campaign.signatures.split()[0]
        if campaign.start_date:
            # Remove the label from the start date
            campaign.start_date = campaign.start_date.split(': ')[-1]
        if campaign.end_date:
            # Remove the label from the end date
            campaign.end_date = campaign.end_date.split(': ')[-1]
        if campaign.organizer:
            # Remove the label from the organizer
            campaign.organizer = campaign.organizer.split(': ')[-1]
        if campaign.url:
            # Add the base URL to the campaign URL
            campaign.url = 'https://www.karzar.net' + campaign.url
        
        # Add the campaign to the list of campaigns
        campaigns.append(campaign)
    
    # Return the list of all campaigns found on this page
    return campaigns

# Function to convert a list of Campaign objects to a pandas DataFrame
def campaigns_to_dataframe(campaigns: List[Campaign]) -> pd.DataFrame:
    # Convert each Campaign object to a dictionary and create a DataFrame
    return pd.DataFrame([vars(campaign) for campaign in campaigns])

# Function to extract province URLs from the HTML content
def get_province_urls(html_content: str) -> Dict[str, str]:
    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')
    # Find all option elements that have a value starting with the campaign URL
    options = soup.select('option[value^="https://www.karzar.net/campaigns?province="]')
    
    # Dictionary to store province names and their URLs
    province_urls = {}
    for option in options:
        # Extract the URL from the 'value' attribute
        url = option['value']
        # Extract the province name from the text content of the option
        province_name = option.text.strip()
        # Add the province and its URL to the dictionary
        province_urls[province_name] = url
    
    # Return the dictionary of province URLs
    return province_urls

# Function to scrape all pages for a single province
def scrape_province_with_pagination(base_url: str, province: str) -> List[Campaign]:
    # List to store all campaigns from all pages
    all_campaigns = []
    # Start with page 1
    page = 1
    
    while True:
        # Construct the URL for the current page
        url = f"{base_url}&page={page}"
        print(f"Scraping {url}")
        # Scrape campaigns from the current page
        campaigns = scrape_karzar_campaigns(url, province)
        
        # If no campaigns are found, we've reached the end of the pages
        if not campaigns:
            break
        
        # Add the campaigns from this page to the list of all campaigns
        all_campaigns.extend(campaigns)
        # Move to the next page
        page += 1
        # Add a 1-second delay between requests to be polite to the server
        time.sleep(1)
    
    # Return the list of all campaigns from all pages
    return all_campaigns

# Function to scrape data for all provinces
def scrape_all_provinces(province_urls: Dict[str, str]) -> Dict[str, pd.DataFrame]:
    # Dictionary to store DataFrames for each province
    province_dataframes = {}
    
    # Iterate through each province and its URL
    for province_name, url in province_urls.items():
        print(f"Scraping data for {province_name}...")
        # Scrape all campaigns for this province
        campaigns = scrape_province_with_pagination(url, province_name)
        # Convert the campaigns to a DataFrame
        df = campaigns_to_dataframe(campaigns)
        # Store the DataFrame in the dictionary
        province_dataframes[province_name] = df
    
    # Return the dictionary of all province DataFrames
    return province_dataframes

# Example usage
html_content = """
 <option value="https://www.karzar.net/campaigns?province=1" >ملی(غیراستانی)</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=2"  selected > آذربايجان شرقی</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=3" >آذربايجان غربی</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=4" >اردبیل</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=5" >اصفهان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=32" >البرز</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=6" >ايلام</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=7" >بوشهر</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=8" >تهران</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=9" >چهارمحال و بختياری</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=10" >خراسان جنوبی</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=11" >خراسان رضوی</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=12" >خراسان شمالی</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=13" >خوزستان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=14" >زنجان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=15" >سمنان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=16" >سيستان و بلوچستان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=17" >فارس</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=18" >قزوين</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=19" >قم</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=20" >كردستان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=21" >كرمان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=22" >كرمانشاه</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=23" >كهكيلويه و بويراحمد</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=24" >گلستان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=25" >گیلان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=26" >لرستان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=27" >مازندران</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=28" >مركزی</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=29" >هرمزگان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=30" >همدان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=31" >یزد</option>
"""

try:
    # Extract province URLs from the HTML content
    province_urls = get_province_urls(html_content)
    
    # Scrape data for all provinces
    province_dataframes = scrape_all_provinces(province_urls)

    # Create "Active Campaigns - Master" DataFrame by concatenating all province DataFrames
    master_df = pd.concat(province_dataframes.values(), ignore_index=True)
    # Add an index column to the master DataFrame
    master_df.insert(0, 'index', range(1, len(master_df) + 1))

    # Print summary of scraped data
    for province_name, df in province_dataframes.items():
        print(f"\n{province_name}:")
        print(f"Number of campaigns: {len(df)}")
        print(df.head())  # Print first 5 rows of each DataFrame

    print("\nActive Campaigns - Master:")
    print(f"Total number of campaigns: {len(master_df)}")
    print(master_df.head())

    # Save each province DataFrame and the master DataFrame to CSV files
    for province_name, df in province_dataframes.items():
        # Create a safe filename by replacing non-alphanumeric characters
        safe_filename = re.sub(r'[^\w\-_\. ]', '_', province_name)
        # Add an index column to the DataFrame
        df.insert(0, 'index', range(1, len(df) + 1))
        # Save the DataFrame to a CSV file
        #df.to_csv(f'{safe_filename}_campaigns.csv', index=False, encoding='utf-8-sig')
        #print(f"Saved {safe_filename}_campaigns.csv")

    # Save the master DataFrame to a CSV file
    master_df.to_csv('Active_Campaigns_Master.csv', index=False, encoding='utf-8-sig')
    print("Saved Active_Campaigns_Master.csv")

# Error handling
except requests.RequestException as e:
    # Handle errors related to making HTTP requests
    print(f"An error occurred while fetching the URL: {e}")
except Exception as e:
    # Handle any other unexpected errors
    print(f"An error occurred while parsing the content: {e}")
```

```python
# Import necessary libraries
import requests  # For making HTTP requests to web pages
from bs4 import BeautifulSoup  # For parsing HTML content
from dataclasses import dataclass  # For creating a data class to store campaign information
from typing import List, Optional, Dict  # For type hinting, making the code more readable and maintainable
import pandas as pd  # For creating and manipulating dataframes
import re  # For using regular expressions (used in creating safe filenames)
import time  # For adding delays between requests

# Define a data class to store information about each campaign
@dataclass
class Campaign:
    # Each field is defined with a type hint and a default value of None
    name: Optional[str] = None  # The name of the campaign
    url: Optional[str] = None  # The URL of the campaign
    signatures: Optional[str] = None  # The number of signatures for the campaign
    start_date: Optional[str] = None  # The start date of the campaign
    end_date: Optional[str] = None  # The end date of the campaign
    organizer: Optional[str] = None  # The organizer of the campaign
    province: Optional[str] = None  # The province where the campaign is active
    archived: bool = True  # Indicates whether the campaign is archived (default is True for this script)

# Function to fetch HTML content from a given URL
def fetch_html_content(url: str) -> str:
    # Send a GET request to the URL
    response = requests.get(url)
    # Raise an exception for bad status codes (e.g., 404 Not Found)
    response.raise_for_status()
    # Return the HTML content as text
    return response.text

# Function to safely extract data from HTML elements
def safe_extract(element, selector, attribute=None):
    # Find the first element matching the CSS selector
    found = element.select_one(selector)
    if found:
        # If an attribute is specified, return its value
        if attribute:
            return found.get(attribute)
        # Otherwise, return the text content of the element, with whitespace stripped
        else:
            return found.get_text(strip=True)
    # If no element is found, return None
    return None

# Function to scrape campaign data from a single page
def scrape_karzar_campaigns(url: str, province: str) -> List[Campaign]:
    # Fetch the HTML content of the page
    html_content = fetch_html_content(url)
    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')
    # Find all campaign box elements on the page
    campaign_boxes = soup.select('div.campaign-box-item')
    
    # List to store all campaigns found on this page
    campaigns = []
    
    # Iterate through each campaign box
    for box in campaign_boxes:
        # Create a Campaign object for each campaign box
        campaign = Campaign(
            # Extract each piece of information using the safe_extract function
            name=safe_extract(box, 'h3.campaign-box__title'),
            url=safe_extract(box, 'a.campaign-box', 'href'),
            signatures=safe_extract(box, 'span.campaign-box__top-text'),
            start_date=safe_extract(box, 'div.campaign-box__info:contains("شروع کارزار")'),
            end_date=safe_extract(box, 'div.campaign-box__info:contains("پایان کارزار")'),
            organizer=safe_extract(box, 'div.campaign-box__info:contains("از طرف")'),
            province=province,
            archived=True  # Set archived to True for all campaigns in this script
        )
        
        # Clean up the extracted data
        if campaign.signatures:
            # Keep only the number part of the signatures
            campaign.signatures = campaign.signatures.split()[0]
        if campaign.start_date:
            # Remove the label from the start date
            campaign.start_date = campaign.start_date.split(': ')[-1]
        if campaign.end_date:
            # Remove the label from the end date
            campaign.end_date = campaign.end_date.split(': ')[-1]
        if campaign.organizer:
            # Remove the label from the organizer
            campaign.organizer = campaign.organizer.split(': ')[-1]
        if campaign.url:
            # Add the base URL to the campaign URL
            campaign.url = 'https://www.karzar.net' + campaign.url
        
        # Add the campaign to the list of campaigns
        campaigns.append(campaign)
    
    # Return the list of all campaigns found on this page
    return campaigns

# Function to convert a list of Campaign objects to a pandas DataFrame
def campaigns_to_dataframe(campaigns: List[Campaign]) -> pd.DataFrame:
    # Convert each Campaign object to a dictionary and create a DataFrame
    return pd.DataFrame([vars(campaign) for campaign in campaigns])

# Function to extract province URLs from the HTML content
def get_province_urls(html_content: str) -> Dict[str, str]:
    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')
    # Find all option elements that have a value starting with the campaign URL
    options = soup.select('option[value^="https://www.karzar.net/campaigns?province="]')
    
    # Dictionary to store province names and their URLs
    province_urls = {}
    for option in options:
        # Extract the URL from the 'value' attribute
        url = option['value']
        # Modify the URL to include the archived parameter
        url = url + '&archived=1'
        # Extract the province name from the text content of the option
        province_name = option.text.strip()
        # Add the province and its modified URL to the dictionary
        province_urls[province_name] = url
    
    # Return the dictionary of province URLs
    return province_urls

# Function to scrape all pages for a single province
def scrape_province_with_pagination(base_url: str, province: str) -> List[Campaign]:
    # List to store all campaigns from all pages
    all_campaigns = []
    # Start with page 1
    page = 1
    
    while True:
        # Construct the URL for the current page
        url = f"{base_url}&page={page}"
        print(f"Scraping {url}")
        # Scrape campaigns from the current page
        campaigns = scrape_karzar_campaigns(url, province)
        
        # If no campaigns are found, we've reached the end of the pages
        if not campaigns:
            break
        
        # Add the campaigns from this page to the list of all campaigns
        all_campaigns.extend(campaigns)
        # Move to the next page
        page += 1
        # Add a 1-second delay between requests to be polite to the server
        time.sleep(1)
    
    # Return the list of all campaigns from all pages
    return all_campaigns

# Function to scrape data for all provinces
def scrape_all_provinces(province_urls: Dict[str, str]) -> Dict[str, pd.DataFrame]:
    # Dictionary to store DataFrames for each province
    province_dataframes = {}
    
    # Iterate through each province and its URL
    for province_name, url in province_urls.items():
        print(f"Scraping archived data for {province_name}...")
        # Scrape all campaigns for this province
        campaigns = scrape_province_with_pagination(url, province_name)
        # Convert the campaigns to a DataFrame
        df = campaigns_to_dataframe(campaigns)
        # Store the DataFrame in the dictionary
        province_dataframes[province_name] = df
    
    # Return the dictionary of all province DataFrames
    return province_dataframes

# Main execution block
try:
    # HTML content containing province options (you would typically fetch this from a website)
    html_content = """
                     <option value="https://www.karzar.net/campaigns?province=1" >ملی(غیراستانی)</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=2"  selected > آذربايجان شرقی</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=3" >آذربايجان غربی</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=4" >اردبیل</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=5" >اصفهان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=32" >البرز</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=6" >ايلام</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=7" >بوشهر</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=8" >تهران</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=9" >چهارمحال و بختياری</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=10" >خراسان جنوبی</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=11" >خراسان رضوی</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=12" >خراسان شمالی</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=13" >خوزستان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=14" >زنجان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=15" >سمنان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=16" >سيستان و بلوچستان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=17" >فارس</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=18" >قزوين</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=19" >قم</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=20" >كردستان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=21" >كرمان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=22" >كرمانشاه</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=23" >كهكيلويه و بويراحمد</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=24" >گلستان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=25" >گیلان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=26" >لرستان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=27" >مازندران</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=28" >مركزی</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=29" >هرمزگان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=30" >همدان</option>
                                    
                    <option value="https://www.karzar.net/campaigns?province=31" >یزد</option>
    """

    # Extract province URLs from the HTML content
    province_urls = get_province_urls(html_content)
    
    # Scrape archived data for all provinces
    province_dataframes = scrape_all_provinces(province_urls)

    # Create "Archived Campaigns - Master" DataFrame by concatenating all province DataFrames
    master_df = pd.concat(province_dataframes.values(), ignore_index=True)
    # Add an index column to the master DataFrame
    master_df.insert(0, 'index', range(1, len(master_df) + 1))

    # Print summary of scraped data
    for province_name, df in province_dataframes.items():
        print(f"\n{province_name} (Archived):")
        print(f"Number of archived campaigns: {len(df)}")
        print(df.head())  # Print first 5 rows of each DataFrame

    print("\nArchived Campaigns - Master:")
    print(f"Total number of archived campaigns: {len(master_df)}")
    print(master_df.head())

    # Save each province DataFrame and the master DataFrame to CSV files
    for province_name, df in province_dataframes.items():
        # Create a safe filename by replacing non-alphanumeric characters
        safe_filename = re.sub(r'[^\w\-_\. ]', '_', province_name)
        # Add an index column to the DataFrame
        df.insert(0, 'index', range(1, len(df) + 1))
        # Save the DataFrame to a CSV file
        #df.to_csv(f'{safe_filename}_archived_campaigns.csv', index=False, encoding='utf-8-sig')
        #print(f"Saved {safe_filename}_archived_campaigns.csv")

    # Save the master DataFrame to a CSV file
    master_df.to_csv('Archived_Campaigns_Master.csv', index=False, encoding='utf-8-sig')
    print("Saved Archived_Campaigns_Master.csv")

# Error handling
except requests.RequestException as e:
    # Handle errors related to making HTTP requests
    print(f"An error occurred while fetching the URL: {e}")
except Exception as e:
    # Handle any other unexpected errors
    print(f"An error occurred while parsing the content: {e}")
```

```python
import requests
from bs4 import BeautifulSoup
from dataclasses import dataclass
from typing import List, Optional, Dict
import pandas as pd
import re
import time
import logging

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

@dataclass
class Campaign:
    name: Optional[str] = None
    url: Optional[str] = None
    signatures: Optional[str] = None
    start_date: Optional[str] = None
    end_date: Optional[str] = None
    organizer: Optional[str] = None
    province: Optional[str] = None
    category: Optional[str] = None

def fetch_html_content(url: str) -> str:
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text
    except requests.RequestException as e:
        logger.error(f"Failed to fetch URL {url}: {e}")
        raise

def safe_extract(element, selector, attribute=None):
    try:
        found = element.select_one(selector)
        if found:
            if attribute:
                return found.get(attribute)
            else:
                return found.get_text(strip=True)
    except Exception as e:
        logger.warning(f"Failed to extract data with selector '{selector}': {e}")
    return None

def scrape_karzar_campaigns(url: str, province: str, category: str) -> List[Campaign]:
    try:
        html_content = fetch_html_content(url)
        soup = BeautifulSoup(html_content, 'html.parser')
        campaign_boxes = soup.select('div.campaign-box-item')
        
        campaigns = []
        
        for box in campaign_boxes:
            campaign = Campaign(
                name=safe_extract(box, 'h3.campaign-box__title'),
                url=safe_extract(box, 'a.campaign-box', 'href'),
                signatures=safe_extract(box, 'span.campaign-box__top-text'),
                start_date=safe_extract(box, 'div.campaign-box__info:contains("شروع کارزار")'),
                end_date=safe_extract(box, 'div.campaign-box__info:contains("پایان کارزار")'),
                organizer=safe_extract(box, 'div.campaign-box__info:contains("از طرف")'),
                province=province,
                category=category
            )
            
            if campaign.signatures:
                campaign.signatures = campaign.signatures.split()[0]
            if campaign.start_date:
                campaign.start_date = campaign.start_date.split(': ')[-1]
            if campaign.end_date:
                campaign.end_date = campaign.end_date.split(': ')[-1]
            if campaign.organizer:
                campaign.organizer = campaign.organizer.split(': ')[-1]
            if campaign.url:
                campaign.url = 'https://www.karzar.net' + campaign.url
            
            campaigns.append(campaign)
        
        return campaigns
    except Exception as e:
        logger.error(f"Failed to scrape campaigns from {url}: {e}")
        return []

def campaigns_to_dataframe(campaigns: List[Campaign]) -> pd.DataFrame:
    return pd.DataFrame([vars(campaign) for campaign in campaigns])

def get_province_urls() -> Dict[str, str]:
    base_url = "https://www.karzar.net/campaigns"
    try:
        html_content = fetch_html_content(base_url)
        soup = BeautifulSoup(html_content, 'html.parser')
        options = soup.select('option[value^="https://www.karzar.net/campaigns?province="]')
        
        province_urls = {}
        for option in options:
            url = option['value']
            province_name = option.text.strip()
            province_id = url.split('=')[-1]
            province_urls[province_name] = province_id
        
        if not province_urls:
            logger.warning("No province URLs were extracted. The page structure might have changed.")
        
        return province_urls
    except Exception as e:
        logger.error(f"Failed to get province URLs: {e}")
        return {}

def get_category_urls() -> Dict[str, str]:
    base_url = "https://www.karzar.net/campaigns"
    try:
        html_content = fetch_html_content(base_url)
        soup = BeautifulSoup(html_content, 'html.parser')
        category_links = soup.select('a[href^="https://www.karzar.net/campaigns?category="]')
        
        category_urls = {}
        for link in category_links:
            try:
                url = link['href']
                category_name = link.get('title') or link.text.strip()
                category_id = url.split('=')[-1]
                category_urls[category_name] = category_id
            except Exception as e:
                logger.warning(f"Failed to process category link: {link}. Error: {e}")
        
        if not category_urls:
            logger.warning("No category URLs were extracted. The page structure might have changed.")
        
        return category_urls
    except Exception as e:
        logger.error(f"Failed to get category URLs: {e}")
        return {}

def scrape_with_pagination(base_url: str, province: str, category: str) -> List[Campaign]:
    all_campaigns = []
    page = 1
    
    while True:
        url = f"{base_url}&page={page}"
        logger.info(f"Scraping {url}")
        campaigns = scrape_karzar_campaigns(url, province, category)
        
        if not campaigns:
            break
        
        all_campaigns.extend(campaigns)
        page += 1
        time.sleep(1)
    
    return all_campaigns

def scrape_all_provinces_and_categories(province_urls: Dict[str, str], category_urls: Dict[str, str]) -> Dict[str, pd.DataFrame]:
    all_dataframes = {}
    
    for province_name, province_id in province_urls.items():
        for category_name, category_id in category_urls.items():
            logger.info(f"Scraping data for {province_name} - {category_name}...")
            base_url = f"https://www.karzar.net/campaigns?category={category_id}&province={province_id}"
            campaigns = scrape_with_pagination(base_url, province_name, category_name)
            df = campaigns_to_dataframe(campaigns)
            key = f"{province_name} - {category_name}"
            all_dataframes[key] = df
    
    return all_dataframes

def main():
    try:
        province_urls = get_province_urls()
        category_urls = get_category_urls()
        
        if not province_urls or not category_urls:
            logger.error("Failed to retrieve province or category URLs. Exiting.")
            return

        all_dataframes = scrape_all_provinces_and_categories(province_urls, category_urls)

        master_df = pd.concat(all_dataframes.values(), ignore_index=True)
        master_df.insert(0, 'index', range(1, len(master_df) + 1))

        for key, df in all_dataframes.items():
            logger.info(f"\n{key}:")
            logger.info(f"Number of campaigns: {len(df)}")
            logger.info(df.head().to_string())

        logger.info("\nActive Campaigns - Master:")
        logger.info(f"Total number of campaigns: {len(master_df)}")
        logger.info(master_df.head().to_string())

        for key, df in all_dataframes.items():
            safe_filename = re.sub(r'[^\w\-_\. ]', '_', key)
            df.insert(0, 'index', range(1, len(df) + 1))
            # Uncomment the following line to save individual CSV files
            # df.to_csv(f'{safe_filename}_campaigns.csv', index=False, encoding='utf-8-sig')
            # logger.info(f"Saved {safe_filename}_campaigns.csv")

        #master_df.to_csv('Active_Campaigns_Master.csv', index=False, encoding='utf-8-sig')
        logger.info("Saved Active_Campaigns_Master.csv")

    except Exception as e:
        logger.error(f"An error occurred during the scraping process: {e}")

if __name__ == "__main__":
    main()
```

```python
import requests
from bs4 import BeautifulSoup
from dataclasses import dataclass
from typing import List, Optional, Dict
import pandas as pd
import re
import time
import logging

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

@dataclass
class Campaign:
    name: Optional[str] = None
    url: Optional[str] = None
    signatures: Optional[str] = None
    start_date: Optional[str] = None
    end_date: Optional[str] = None
    organizer: Optional[str] = None
    province: Optional[str] = None
    category: Optional[str] = None

def fetch_html_content(url: str) -> str:
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text
    except requests.RequestException as e:
        logger.error(f"Failed to fetch URL {url}: {e}")
        raise

def safe_extract(element, selector, attribute=None):
    try:
        found = element.select_one(selector)
        if found:
            if attribute:
                return found.get(attribute)
            else:
                return found.get_text(strip=True)
    except Exception as e:
        logger.warning(f"Failed to extract data with selector '{selector}': {e}")
    return None

def scrape_karzar_campaigns(url: str, province: str, category: str) -> List[Campaign]:
    try:
        html_content = fetch_html_content(url)
        soup = BeautifulSoup(html_content, 'html.parser')
        campaign_boxes = soup.select('div.campaign-box-item')
        
        campaigns = []
        
        for box in campaign_boxes:
            campaign = Campaign(
                name=safe_extract(box, 'h3.campaign-box__title'),
                url=safe_extract(box, 'a.campaign-box', 'href'),
                signatures=safe_extract(box, 'span.campaign-box__top-text'),
                start_date=safe_extract(box, 'div.campaign-box__info:contains("شروع کارزار")'),
                end_date=safe_extract(box, 'div.campaign-box__info:contains("پایان کارزار")'),
                organizer=safe_extract(box, 'div.campaign-box__info:contains("از طرف")'),
                province=province,
                category=category
            )
            
            if campaign.signatures:
                campaign.signatures = campaign.signatures.split()[0]
            if campaign.start_date:
                campaign.start_date = campaign.start_date.split(': ')[-1]
            if campaign.end_date:
                campaign.end_date = campaign.end_date.split(': ')[-1]
            if campaign.organizer:
                campaign.organizer = campaign.organizer.split(': ')[-1]
            if campaign.url:
                campaign.url = 'https://www.karzar.net' + campaign.url
            
            campaigns.append(campaign)
        
        return campaigns
    except Exception as e:
        logger.error(f"Failed to scrape campaigns from {url}: {e}")
        return []

def campaigns_to_dataframe(campaigns: List[Campaign]) -> pd.DataFrame:
    return pd.DataFrame([vars(campaign) for campaign in campaigns])

def get_province_urls() -> Dict[str, str]:
    base_url = "https://www.karzar.net/campaigns"
    try:
        html_content = fetch_html_content(base_url)
        soup = BeautifulSoup(html_content, 'html.parser')
        options = soup.select('option[value^="https://www.karzar.net/campaigns?province="]')
        
        province_urls = {}
        for option in options:
            url = option['value']
            province_name = option.text.strip()
            province_id = url.split('=')[-1]
            province_urls[province_name] = province_id
        
        if not province_urls:
            logger.warning("No province URLs were extracted. The page structure might have changed.")
        
        return province_urls
    except Exception as e:
        logger.error(f"Failed to get province URLs: {e}")
        return {}

def get_category_urls() -> Dict[str, str]:
    base_url = "https://www.karzar.net/campaigns"
    try:
        html_content = fetch_html_content(base_url)
        soup = BeautifulSoup(html_content, 'html.parser')
        category_links = soup.select('a[href^="https://www.karzar.net/campaigns?category="]')
        
        category_urls = {}
        for link in category_links:
            try:
                url = link['href']
                category_name = link.get('title') or link.text.strip()
                category_id = url.split('=')[-1]
                category_urls[category_name] = category_id
            except Exception as e:
                logger.warning(f"Failed to process category link: {link}. Error: {e}")
        
        if not category_urls:
            logger.warning("No category URLs were extracted. The page structure might have changed.")
        
        return category_urls
    except Exception as e:
        logger.error(f"Failed to get category URLs: {e}")
        return {}

def scrape_with_pagination(base_url: str, province: str, category: str) -> List[Campaign]:
    all_campaigns = []
    page = 1
    
    while True:
        url = f"{base_url}&page={page}"
        logger.info(f"Scraping {url}")
        campaigns = scrape_karzar_campaigns(url, province, category)
        
        if not campaigns:
            break
        
        all_campaigns.extend(campaigns)
        page += 1
        time.sleep(1)
    
    return all_campaigns

def scrape_all_provinces_and_categories(province_urls: Dict[str, str], category_urls: Dict[str, str]) -> Dict[str, pd.DataFrame]:
    all_dataframes = {}
    
    for province_name, province_id in province_urls.items():
        for category_name, category_id in category_urls.items():
            logger.info(f"Scraping data for {province_name} - {category_name}...")
            base_url = f"https://www.karzar.net/campaigns?category={category_id}&province={province_id}&archived=1"
            campaigns = scrape_with_pagination(base_url, province_name, category_name)
            df = campaigns_to_dataframe(campaigns)
            key = f"{province_name} - {category_name}"
            all_dataframes[key] = df
    
    return all_dataframes

def main():
    try:
        province_urls = get_province_urls()
        category_urls = get_category_urls()
        
        if not province_urls or not category_urls:
            logger.error("Failed to retrieve province or category URLs. Exiting.")
            return

        all_dataframes = scrape_all_provinces_and_categories(province_urls, category_urls)

        master_df = pd.concat(all_dataframes.values(), ignore_index=True)
        master_df.insert(0, 'index', range(1, len(master_df) + 1))

        for key, df in all_dataframes.items():
            logger.info(f"\n{key}:")
            logger.info(f"Number of campaigns: {len(df)}")
            logger.info(df.head().to_string())

        logger.info("\nArchived Campaigns - Master:")
        logger.info(f"Total number of campaigns: {len(master_df)}")
        logger.info(master_df.head().to_string())

        for key, df in all_dataframes.items():
            safe_filename = re.sub(r'[^\w\-_\. ]', '_', key)
            df.insert(0, 'index', range(1, len(df) + 1))
            # Uncomment the following line to save individual CSV files
            # df.to_csv(f'{safe_filename}_archived_campaigns.csv', index=False, encoding='utf-8-sig')
            # logger.info(f"Saved {safe_filename}_archived_campaigns.csv")

        master_df.to_csv('Archived_Campaigns_Master.csv', index=False, encoding='utf-8-sig')
        logger.info("Saved Archived_Campaigns_Master.csv")

    except Exception as e:
        logger.error(f"An error occurred during the scraping process: {e}")

if __name__ == "__main__":
    main()
```

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import time
import random
import os

# Function to clean text by removing extra whitespace
def clean_text(text):
    return re.sub(r'\s+', ' ', text).strip()

# Function to scrape campaign content
def scrape_campaign_content(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract the title
    campaign_title = soup.title.string if soup.title else "No title found"

    # Extract the description from the meta tag
    description = soup.find('meta', attrs={'name': 'description'})
    description_text = description['content'] if description else "No description found"

    # Extract addressee and addressee title from the full text content
    fulltext_div = soup.find('div', class_='fulltext')
    addressee_text = addressee_title_text = full_text = "No full text found"
    if fulltext_div:
        addressee = fulltext_div.find('p', string=lambda s: s and 'جناب آقای' in s)
        addressee_title = addressee.find_next('p') if addressee else None
        addressee_text = clean_text(addressee.text) if addressee else "No addressee found"
        addressee_title_text = clean_text(addressee_title.text) if addressee_title else "No addressee title found"

        full_text_list = [clean_text(p.text) for p in fulltext_div.find_all('p') if p != addressee and p != addressee_title]
        full_text = " ".join(full_text_list)

    # Extract additional metadata dynamically
    metadata_items_dict = {}
    metadata_div = soup.find('div', class_='metadata')
    if metadata_div:
        metadata_items = metadata_div.find_all('div', class_='col-lg-6')
        for item in metadata_items:
            cleaned_item = clean_text(item.text)
            if 'شروع کارزار' in cleaned_item:
                metadata_items_dict['start_of_campaign'] = cleaned_item.split(':')[-1].strip()
            elif 'پایان کارزار' in cleaned_item:
                metadata_items_dict['end_of_campaign'] = cleaned_item.split(':')[-1].strip()
            elif 'از طرف' in cleaned_item:
                metadata_items_dict['from_info'] = cleaned_item.split(':')[-1].strip()
            elif 'هشتگ رسمی' in cleaned_item:
                metadata_items_dict['official_hashtag'] = cleaned_item.split(':')[-1].strip()
            elif 'نویسنده' in cleaned_item:
                metadata_items_dict['author'] = cleaned_item.split(':')[-1].strip()
            elif 'لینک کارزار' in cleaned_item:
                metadata_items_dict['campaign_link'] = cleaned_item.split(':')[-1].strip()

    # Extract progress information
    progress_bar = soup.find('div', class_='progress-bar')
    progress_percentage = progress_bar.text.strip() if progress_bar else "No progress found"

    # Extract remaining time or other information
    remaining_text = soup.find('span', class_='fa')
    remaining_info = clean_text(remaining_text.text) if remaining_text else "No remaining info found"

    # Return the collected data
    return pd.DataFrame({
        'campaign_title': [campaign_title],
        'description': [description_text],
        'addressee': [addressee_text],
        'addressee_title': [addressee_title_text],
        'full_text': [full_text],
        'start_of_campaign': [metadata_items_dict.get('start_of_campaign', 'No start date found')],
        'end_of_campaign': [metadata_items_dict.get('end_of_campaign', 'No end date found')],
        'from_info': [metadata_items_dict.get('from_info', 'No from info found')],
        'official_hashtag': [metadata_items_dict.get('official_hashtag', 'No official hashtag found')],
        'author': [metadata_items_dict.get('author', 'No author found')],
        'progress_percentage': [progress_percentage],
        'remaining_info': [remaining_info],
        'campaign_link': [url],
    })

# Function to scrape signature counts
def count_signatures(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    signature_elements = soup.select('.supporters ul li')
    return len(signature_elements)

# Function to scrape comments
def scrape_comments(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    names, comments = [], []
    comment_blocks = soup.find_all('div', class_='comment')

    for comment in comment_blocks:
        name = comment.find('span', class_='comment-name').text.strip().split()[0]
        comment_text = comment.find('b').next_sibling.strip() if comment.find('b').next_sibling else ''
        names.append(name)
        comments.append(comment_text)

    return pd.DataFrame({
        'name': names,
        'comment': comments,
        'url': [url] * len(names)
    })

# Function to scrape media attention
def scrape_media_updates(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    media_names, headlines, links = [], [], []
    media_updates = soup.find_all('div', class_='updates-news')

    for update in media_updates:
        img_tag = update.find('img')
        media_name = img_tag['alt'] if img_tag and 'alt' in img_tag.attrs else 'Unknown Media'
        headline = update.find('h4').text.strip()
        link = update.find('a')['href']
        media_names.append(media_name)
        headlines.append(headline)
        links.append(link)

    return pd.DataFrame({
        'media_name': media_names,
        'headline': headlines,
        'link': links,
        'url': [url] * len(media_names)
    })

# Define the updated_master_df globally so it's accessible after stopping the code
updated_master_df = None

def process_master_data(master_df, save_file_path="updated_master_data.xlsx"):
    global updated_master_df  # Make sure it's available globally
    all_content_data = []
    all_signature_data = []
    all_comment_data = []
    all_media_data = []

    # If the file already exists, load it to resume progress
    if os.path.exists(save_file_path):
        updated_master_df = pd.read_excel(save_file_path)
        processed_urls = updated_master_df['url'].unique().tolist()
    else:
        updated_master_df = master_df.copy()
        processed_urls = []

    for index, row in master_df.iterrows():
        index = row['index']
        url = row['url']

        # Skip URLs that have already been processed
        if url in processed_urls:
            print(f"Skipping already processed URL: {url}")
            continue

        try:
            print(f"Index: {index}")
            print(f"Processing URL: {url}")

            # Scrape the different types of data
            content_data = scrape_campaign_content(url)
            signature_data = pd.DataFrame({'url': [url], 'signatures_count': [count_signatures(url)]})
            comment_data = scrape_comments(url)
            media_data = scrape_media_updates(url)

            # Merge data for this URL with the master DataFrame
            updated_master_df = updated_master_df.merge(content_data, on='url', how='left')
            updated_master_df = updated_master_df.merge(signature_data, on='url', how='left')
            updated_master_df = updated_master_df.merge(comment_data, on='url', how='left')
            updated_master_df = updated_master_df.merge(media_data, on='url', how='left')

            # Save progress after each URL
            updated_master_df.to_excel(save_file_path, index=False)
            print(f"Saved progress after processing {url}")

            # Pause to avoid overloading the server
            time.sleep(random.uniform(1, 3))

        except Exception as e:
            #print(f"Error processing URL {url}: {e}")

    return updated_master_df

# Load the master DataFrame
master_df = pd.read_excel("C:\\Users\\Saber\\OneDrive\\Desktop\\Ali Kadivar\\Karzar\\Karzar Master Data - 2024_9_9.xlsx")  # Explicitly specify the engine

# Process the master DataFrame
updated_master_df = process_master_data(master_df, save_file_path="updated_master_data.xlsx")

print("Data collection and merging completed.")
```

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import time
import random
import os

def clean_text(text):
    return re.sub(r'\s+', ' ', text).strip()

def scrape_campaign_content(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    campaign_title = soup.title.string if soup.title else "No title found"
    description = soup.find('meta', attrs={'name': 'description'})
    description_text = description['content'] if description else "No description found"

    fulltext_div = soup.find('div', class_='fulltext')
    full_text = "No full text found"
    if fulltext_div:
        full_text_list = [clean_text(p.text) for p in fulltext_div.find_all('p')]
        full_text = " ".join(full_text_list)

    metadata_items_dict = {}
    metadata_div = soup.find('div', class_='metadata')
    if metadata_div:
        metadata_items = metadata_div.find_all('div', class_='col-lg-6')
        for item in metadata_items:
            cleaned_item = clean_text(item.text)
            if 'شروع کارزار' in cleaned_item:
                metadata_items_dict['start_of_campaign'] = cleaned_item.split(':')[-1].strip()
            elif 'پایان کارزار' in cleaned_item:
                metadata_items_dict['end_of_campaign'] = cleaned_item.split(':')[-1].strip()
            elif 'از طرف' in cleaned_item:
                metadata_items_dict['from_info'] = cleaned_item.split(':')[-1].strip()
            elif 'هشتگ رسمی' in cleaned_item:
                metadata_items_dict['official_hashtag'] = cleaned_item.split(':')[-1].strip()
            elif 'نویسنده' in cleaned_item:
                metadata_items_dict['author'] = cleaned_item.split(':')[-1].strip()
            elif 'لینک کارزار' in cleaned_item:
                metadata_items_dict['campaign_link'] = cleaned_item.split(':')[-1].strip()

    progress_bar = soup.find('div', class_='progress-bar')
    progress_percentage = progress_bar.text.strip() if progress_bar else "No progress found"

    remaining_text = soup.find('span', class_='fa')
    remaining_info = clean_text(remaining_text.text) if remaining_text else "No remaining info found"

    return {
        'campaign_title': campaign_title,
        'description': description_text,
        'full_text': full_text,
        'start_of_campaign': metadata_items_dict.get('start_of_campaign', 'No start date found'),
        'end_of_campaign': metadata_items_dict.get('end_of_campaign', 'No end date found'),
        'from_info': metadata_items_dict.get('from_info', 'No from info found'),
        'official_hashtag': metadata_items_dict.get('official_hashtag', 'No official hashtag found'),
        'author': metadata_items_dict.get('author', 'No author found'),
        'progress_percentage': progress_percentage,
        'remaining_info': remaining_info,
        'campaign_link': url,
    }

def get_max_pages(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    pagination = soup.select_one('.pagination')
    
    if pagination:
        page_numbers = [int(link.text) for link in pagination.find_all('a') if link.text.isdigit()]
        return max(page_numbers) if page_numbers else 1
    else:
        return 1

def scrape_signatures(url):
    max_pages = get_max_pages(url)
    signatures = []
    
    for page in range(1, max_pages + 1):
        print(f"Scraping comments from page {page} of {max_pages}!")
        
        page_url = f"{url}?page={page}"
        response = requests.get(page_url)
        soup = BeautifulSoup(response.content, 'html.parser')
        
        signature_elements = soup.select('.supporters ul li')
        
        for element in signature_elements:
            name = element.find('p').text.strip()
            signatures.append(name)
        
        time.sleep(random.uniform(1, 3))
    
    return signatures

def scrape_comments(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    comments = []
    comment_blocks = soup.find_all('div', class_='comment')

    for comment in comment_blocks:
        name = comment.find('span', class_='comment-name').text.strip().split()[0]
        comment_text = comment.find('b').next_sibling.strip() if comment.find('b').next_sibling else ''
        comments.append({'name': name, 'comment': comment_text})

    return comments

def scrape_media_updates(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    media_updates = []
    media_update_elements = soup.find_all('div', class_='updates-news')

    for update in media_update_elements:
        img_tag = update.find('img')
        media_name = img_tag['alt'] if img_tag and 'alt' in img_tag.attrs else 'Unknown Media'
        headline = update.find('h4').text.strip()
        link = update.find('a')['href']
        media_updates.append({'media_name': media_name, 'headline': headline, 'link': link})

    return media_updates

def get_latest_output_file(base_name="updated_master_data"):
    files = [f for f in os.listdir() if f.startswith(base_name) and f.endswith('.xlsx')]
    if not files:
        return f"{base_name}_run1.xlsx"
    else:
        latest_file = max(files, key=os.path.getmtime)
        run_number = int(latest_file.split('_run')[1].split('.xlsx')[0])
        return f"{base_name}_run{run_number + 1}.xlsx"

def process_master_data(master_df):
    output_file = get_latest_output_file()
    new_master_data = []
    last_processed_index = -1

    if os.path.exists(output_file):
        # Load existing data
        existing_df = pd.read_excel(output_file)
        new_master_data = existing_df.to_dict('records')
        processed_urls = set(existing_df['url'])
        
        # Find the last processed URL in the master_df
        for index, row in master_df.iterrows():
            if row['url'] in processed_urls:
                last_processed_index = index
            else:
                break
    else:
        processed_urls = set()

    # Start processing from the URL after the last processed one
    for index, row in master_df.iterrows():
        index = row['index']
        url = row['url']

        # Skip already processed URLs
        if url in processed_urls:
            print(f"Skipping already processed URL: {url}")
            continue

        try:
            print(f"Index: {index}")
            print(f"Processing URL: {url}")

            content_data = scrape_campaign_content(url)
            signatures = scrape_signatures(url)
            comments = scrape_comments(url)
            media_updates = scrape_media_updates(url)

            campaign_data = {
                'url': url,
                'signatures': signatures,
                'comments': comments,
                'media_updates': media_updates,
                **content_data
            }

            new_master_data.append(campaign_data)

            # Save intermediate results after each URL
            pd.DataFrame(new_master_data).to_excel(output_file, index=False)

            print(f"Completed processing {url}")
            time.sleep(random.uniform(1, 3))

        except Exception as e:
            print(f"Error processing URL {url}: {e}")

        # Save intermediate results after each URL, even if there was an error
        pd.DataFrame(new_master_data).to_excel(output_file, index=False)

    return pd.DataFrame(new_master_data)

# Load the master DataFrame
master_df = pd.read_excel("C:\\Users\\Saber\\OneDrive\\Desktop\\Ali Kadivar\\Karzar\\Karzar Master Data - 2024_9_9.xlsx")

# Process the master DataFrame
updated_master_df = process_master_data(master_df)

print("Data collection and processing completed.")
print(f"Results saved in: {get_latest_output_file()}")
```
